```python
tempo = np.random.uniform(0, 10, 50)
```

```python
tempo
```

```python
salario = 1500 + 300 * tempo + np.random.normal(0, 200, 50)
```

```python
salario
```

```python
plt.scatter(tempo, salario, marker='o', color='b')
plt.title('Projeção Salário')
plt.xlabel('Tempo em anos')
plt.ylabel('Salario')
plt.show()
```

```python
dict = {
    "tempo": tempo,
    "salario": salario
}
df = pd.DataFrame(dict)
df.corr()

#quanto mais próximo do 1 positivo, as variáveis tem maior correlação
```

```python
# Regressão Linear

media_x = np.mean(tempo)
media_y = np.mean(salario)

numerador = np.sum((salario - media_y) * (tempo - media_x))
denominador = np.sum((tempo - media_x) ** 2)
coeficiente = numerador / denominador

intercept = media_y - coeficiente * media_x

y_pred = coeficiente * tempo + intercept
y_pred
```

```python
# Aplicação prática
novo_tempo = 9.5
novo_salario = coeficiente * novo_tempo + intercept
novo_salario
```

```python
# Plotagem dos gráficos

plt.scatter(tempo, salario, marker='o', color='b')
plt.scatter(novo_tempo, novo_salario, marker='*', s=600, color='green')
plt.plot(tempo, y_pred, color='red', label='Y_PRED')
plt.title('Tempo')
plt.xlabel('Tempo em anos')
plt.ylabel('Salario')
plt.show()
```
